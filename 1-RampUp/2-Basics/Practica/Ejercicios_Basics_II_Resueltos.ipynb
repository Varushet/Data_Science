## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math

grados = float(input("Introduzca los grados").replace(",","."))
radianes = grados * math.pi / 180
print(radianes)
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
base = float(input("Base: ").replace(',','.'))
altura = float(input("Altura: ").replace(',','.'))

area = round(base * altura, 2)

print(area)
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
valor_1 = float(input("Introduzca primer valor"))
valor_2 = float(input("Introduzca segundo valor"))

suma = valor_1 + valor_2

print("Es igual a 10", suma == 10)
print("Es superior a 10", suma > 10)
print("Es inferior a 10", suma < 10)
````

## Ejercicio 8
Razona sin ejecutar código el output que obtendremos de las siguientes sentencias
1. True and True and False
2. not ((True or False) and (True or False))
3. (False or False or False or True) and False
4. not False and True and not True

Compruébalo después con python

1. False: La condición "and False" no es verdadera.
2. False: El resultado de la condicón "or" es verdadera al ser binaria, 
pero el "not" que precede al primer argumento hace que la condicion "and True" no sea verdadero.
3. False: El resultado de la condicón "or" es verdadera al ser binaria,
por lo que la condicion "and False" es falsa
4. False: La condición "not True" no es verdadera y las demás sí.

````python
print(True and True and False)
print(not ((True or False) and (True or False)))
print((False or False or False or True) and False)
print(not False and True and not True)
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
frase = "Lo primero de todo, cómo están los máquinas"
print(frase.upper())
print(frase.lower())
print(frase.title())
print(frase.split(" "))
print(frase.replace(",", ";"))
print(frase.replace("a", ""))
````
