## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
operaciones = ("mult", "div", "suma", "resta", "pot", "salir")

while True:
    opera = input("Qué operación vas a realizar? ").lower()
    if opera == "salir":
        print("Saliendo...")
        break
    if opera not in operaciones:
        print("Operación no válida, pruebe otra vez")
        continue

    try:
        n1 = float(input())
        n2 = float(input())
    except ValueError:
        print("Entrada no válida, introduce números.")
        continue

    match opera:
        case "mult":
            print(n1 * n2)
        case "div":
            if n2 == 0:
                print("No se puede dividir entre 0.")
            else:
                print(n1 / n2)
        case "suma":
            print(n1 + n2)
        case "resta":
            print(n1 - n2)
        case "pot":
            print(n1 ** n2)
```
